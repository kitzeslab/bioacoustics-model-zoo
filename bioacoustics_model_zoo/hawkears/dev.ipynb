```python
import torch 
from pathlib import Path
```

```python
all_checkpoints = glob('/Users/SML161/HawkEars/data/ckpt/*')
all_models = []
classes = None
class_codes = None

for ckpt_path in all_checkpoints:
    mdict = torch.load(ckpt_path,map_location=torch.device('cpu'))  # Load the first checkpoint

    model_name = mdict['hyper_parameters']['model_name']
    tokens = model_name.split('_')
    m_classes = mdict['hyper_parameters']['train_class_names']
    m_class_codes = mdict['hyper_parameters']['train_class_codes']

    if class_codes is None:
        class_codes = m_class_codes
    else:
        assert class_codes == m_class_codes, "Class codes do not match across models"
    
    if classes is None:
        classes = m_classes
    else:
        assert classes == m_classes, "Class names do not match across models"

    # if a new ensemble of models uses a different architecture, we need to change the model constructor logic
    model = architecture_constructors.get_hgnet(tokens[-1], num_classes=len(classes))
    # remove 'base_model' prefix from state dict keys to match our architecture
    state_dict = {
        k.replace("base_model.", ""): v
        for k, v in mdict["state_dict"].items()
    }
    model.load_state_dict(state_dict)
    all_models.append(model)

    
```

```python
len(all_models)
```

```python
class Ensemble(torch.nn.Module):
    """Ensemble of multiple models for classification

    Args:
        models: list of models to use in the ensemble
    """

    def __init__(self, models):
        super(Ensemble, self).__init__()
        self.models = models
        for i, m in enumerate(models):
            self.add_module(f'model_{i}', m)

    def forward(self, x):
        """forward pass through the ensemble"""
        outputs = [model(x) for model in self.models]
        return torch.mean(torch.stack(outputs), dim=0)
```

```python
ens = Ensemble(all_models)
```

```python
# test with random inputs
ens = ens.to('mps')
ens.eval()

shape = [1,192, 384]
x = torch.rand(10, *shape).to('mps')

with torch.no_grad():
    output = ens(x)
    print(output.shape)
```
